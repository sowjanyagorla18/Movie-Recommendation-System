```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv("/content/imdb_top_1000.csv")
```

```python
movies.head()
```

```python
movies.columns
```

```python
def con(name):
    return name.replace(" ", "")
```

```python
movies['Director']=movies['Director'].apply(con)
movies['Star1']=movies['Star1'].apply(con)
movies['Star2']=movies['Star2'].apply(con)
movies['Star3']=movies['Star3'].apply(con)
movies['Star4']=movies['Star4'].apply(con)
```

```python
movies
```

```python
movies['Overview'] = movies['Overview'].apply(lambda x:x.split())
```

```python
def con1(name):
    return name.replace(",", " ")
movies['Genre']=movies['Genre'].apply(con1)
```

```python
movies
```

```python

movies['Overview'] = movies['Overview'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)
movies['Genre'] = movies['Genre'].astype(str)
movies['Director'] = movies['Director'].astype(str)
movies['Star1'] = movies['Star1'].astype(str)
movies['Star2'] = movies['Star2'].astype(str)
movies['Star3'] = movies['Star3'].astype(str)
movies['Star4'] = movies['Star4'].astype(str)

movies['tags'] = (movies['Overview'] + ' ' + movies['Genre'] + ' ' +
                  movies['Director'] + ' ' + movies['Star1'] + ' ' +
                  movies['Star2'] + ' ' + movies['Star3'] + ' ' + movies['Star4'])


print(movies['tags'])
```

```python
new_movies = movies[['Poster_Link', 'Series_Title', 'tags']]
new_movies.head()
```

```python
new_movies.tags[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_movies['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    index =new_movies[new_movies['Series_Title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_movies.iloc[i[0]].Series_Title)
```

```python
recommend('The Godfather')
```
